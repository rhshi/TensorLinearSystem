```julia
include("testingC.jl");
using Symbolics, Graphs, GraphPlot, Latexify;
```

```julia
n = 5
c = 2;
r = Int(1+(c+1)*n+c*(1-c)/2)

A = randn(n+1, r);

getl, getL, sumL = makeSum(n, c);
```

```julia
X, Y, varTups, eqTups, basis, h = getReducedMat(A, c);
```

```julia
Y
```

```julia
varInds = []
for (i, var) in enumerate(varTups)
    if sum([x for x in var][1:c]) == 3
        push!(varInds, i)
    end 
end
```

```julia
Z2 = Y[:, setdiff(1:size(Y)[2], varInds)];
```

```julia
eqInds = []
for (i, eqTup) in enumerate(eqTups)
    if !(eqTup[1][1] isa Number)
        push!(eqInds, i)
    end
end

Z1_ = Y[eqInds, varInds];
```

```julia
Id = diagm(ones(size(Z1_)[1]));
```

```julia
Z1 = copy(Z1_)
for (i, eqInd) in enumerate(eqInds)
    Z1 = hcat(Z1[:, 1:eqInd-1], -h[1]*Id[:, i], Z1[:, eqInd:end])
end
```
